## Day 7: Camel Cards
### Part 1:

```python
# Open input
with open('input.txt', 'r') as hands:
    hands = [hand.strip() for hand in hands.readlines()]

# Create dictionary of hands and bids
hand_bid_dict = dict(zip([hand.split(" ")[0] for hand in hands],
                         [int(hand.split(" ")[1]) for hand in hands]))

# Define lists to sort the hands in
five_list = []
four_list = []
full_list = []
three_list = []
twop_list = []
onep_list = []
high_list = []

hands_sorted = []

# Determine hand type
for hand in list(hand_bid_dict.keys()):
    
    # Sort hands to group cards together
    sorted_hand = sorted(hand)
    
    # Five of a kind
    if sorted_hand[0] == sorted_hand[4]:
        five_list.append(hand)
    # 4 of a kind
    elif hand.count(sorted_hand[1]) == 4:
        four_list.append(hand)
    # Full house
    elif (hand.count(sorted_hand[0]) == 3 and hand.count(sorted_hand[4]) == 2) or \
    (hand.count(sorted_hand[0]) == 2 and hand.count(sorted_hand[4]) == 3):
        full_list.append(hand)
    # Three of a kind
    elif hand.count(sorted_hand[2]) == 3:
        three_list.append(hand)
    # Two pair
    elif hand.count(sorted_hand[1]) == 2 and hand.count(sorted_hand[3]) == 2:
        twop_list.append(hand)
    # One pair
    elif len(hand) != len(set(hand)): 
        onep_list.append(hand)
    else: 
        high_list.append(hand)

# Determine letter values
letter_values = {"A": "14", "K": "13", "Q": "12", "J": "11", "T": "10"}

# Sort type_list by hand values
for type_list in [high_list, onep_list, twop_list, three_list, full_list, four_list, five_list]:
    for i in range(len(type_list)):
        hand_list = [card for card in type_list[i]]
        # Translate the letters into numbers and turn everything into ints
        for j in range(len(hand_list)):
            if hand_list[j] in letter_values:
                hand_list[j] = int(letter_values[hand_list[j]])
            else:
                hand_list[j] = int(hand_list[j])
        type_list[i] = hand_list

    for hand in sorted(type_list):
        hands_sorted.append(hand)

# Translate back and calculate bid*hand_index
# Initiate counter 
total_winnings = 0

inv_letter_values = {value: letter for letter, value in letter_values.items()}

# Translate back and calculate winnings
for i in range(len(hands_sorted)):
    for j in range(len(hands_sorted[i])):
        if str(hands_sorted[i][j]) in inv_letter_values:
            hands_sorted[i][j] = inv_letter_values[str(hands_sorted[i][j])]
        else: 
            hands_sorted[i][j] = str(hands_sorted[i][j])
    hands_sorted[i] = "".join(hands_sorted[i])
    
    # Add winnings to total
    total_winnings += hand_bid_dict[hands_sorted[i]] * (i + 1)

# Print answer
print(f'The total winnings were {total_winnings}.')
```

```python
# Open input
with open('input.txt', 'r') as hands:
    hands = [hand.strip() for hand in hands.readlines()]

# Create dictionary of hands and bids
hand_bid_dict = dict(zip([hand.split(" ")[0] for hand in hands],
                         [int(hand.split(" ")[1]) for hand in hands]))

# Define lists to sort the hands in
five_list = []
four_list = []
full_list = []
three_list = []
twop_list = []
onep_list = []
high_list = []

hands_sorted = []

# Determine hand type
for hand in list(hand_bid_dict.keys()):

    sorted_hand = sorted(hand)

    # Five of a kind
    if len([c for c in list(set(hand)) if c != "J"]) <= 1:
        five_list.append(hand)
    # All different
    elif len([c for c in list(set(hand)) if c != "J"]) == 5:
        high_list.append(hand)

    # Old rules if no Joker present    
    elif "J" not in hand:
        # 4 of a kind
        if hand.count(sorted_hand[1]) == 4:
            four_list.append(hand)
        # Full house
        elif (hand.count(sorted_hand[0]) == 3 and hand.count(sorted_hand[4]) == 2) or \
        (hand.count(sorted_hand[0]) == 2 and hand.count(sorted_hand[4]) == 3):
            full_list.append(hand)
        # Three of a kind
        elif hand.count(sorted_hand[2]) == 3:
            three_list.append(hand)
        # Two pair
        elif hand.count(sorted_hand[1]) == 2 and hand.count(sorted_hand[3]) == 2:
            twop_list.append(hand)
        # One pair
        elif len(hand) != len(set(hand)): 
            onep_list.append(hand)
    
    # If any Joker present (Two pair impossible)
    else:
        type_found = False
        # 4 of a kind
        for card in hand:
            if card != "J":
                if hand.count(card) + hand.count("J") == 4:
                    type_found = True
                    four_list.append(hand)
                    break
        # Full house
        if not type_found and \
        hand.count("J") == 1 and \
        len([c for c in list(set(hand)) if c != "J"]) == 2:
             full_list.append(hand)
        # Three of a kind
        elif 1 <= hand.count("J") <= 2 and \
        len([c for c in list(set(hand)) if c != "J"]) == 3:
            three_list.append(hand)
        # One pair
        elif len([c for c in list(set(hand)) if c != "J"]) == 4:
            onep_list.append(hand)

letter_values = {"A": "14", "K": "13", "Q": "12", "J": "1", "T": "10"}

# Sort type_list by hand values
for type_list in [high_list, onep_list, twop_list, three_list, full_list, four_list, five_list]:

    for i in range(len(type_list)):
        hand_list = [card for card in type_list[i]]
        # Translate the letters into numbers and turn everything into ints
        for j in range(len(hand_list)):
            if hand_list[j] in letter_values:
                hand_list[j] = int(letter_values[hand_list[j]])
            else:
                hand_list[j] = int(hand_list[j])
        type_list[i] = hand_list

    for hand in sorted(type_list):
        hands_sorted.append(hand)

# Translate back and calculate bid*hand_index
# Initiate counter 
total_winnings = 0

inv_letter_values = {value: letter for letter, value in letter_values.items()}

# Translate back and calculate winnings
for i in range(len(hands_sorted)):
    for j in range(len(hands_sorted[i])):
        if str(hands_sorted[i][j]) in inv_letter_values:
            hands_sorted[i][j] = inv_letter_values[str(hands_sorted[i][j])]
        else: 
            hands_sorted[i][j] = str(hands_sorted[i][j])
    hands_sorted[i] = "".join(hands_sorted[i])
    
    # Add winnings to total
    total_winnings += hand_bid_dict[hands_sorted[i]] * (i + 1)

# Print answer
print(f'The total winnings were {total_winnings}.')
```
