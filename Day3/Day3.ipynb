## Day 3: Gear Ratios
### Part 1:

```python
import regex as re

# Open input file
with open('input.txt', 'r') as schematic:
    schematic = schematic.read().split("\n")[:-1]

# Create no_symbol string
no_symbol = "0123456789."

# Initialize counter
part_counter = 0

# All the middle lines
for line_index in range(1, len(schematic) - 1):

    # Define current, last and next lines
    line = schematic[line_index]
    prev_line = schematic[line_index - 1]
    next_line = schematic[line_index + 1]

    # Find indices of all the numbers
    for number in re.finditer(r"\d+", line):
            
        # Initialize part boolean
        part_number = False

        # Check for adjacent symbols
        if number.start() != 0:
            if line[number.start() - 1] not in no_symbol or \
            set(prev_line[number.start() - 1 : number.end()]) - set(no_symbol) or \
            set(next_line[number.start() - 1 : number.end()]) - set(no_symbol): 
                part_number = True
        if number.end() != len(line):
            if line[number.end()] not in no_symbol or \
            set(prev_line[number.start() : number.end() + 1]) - set(no_symbol) or \
            set(next_line[number.start() : number.end() + 1]) - set(no_symbol): 
                    part_number = True

        # Adjust part_counter
        if part_number:
            part_counter += int(line[number.start() : number.end()])

# Add first and last lines:
# First line 
line = schematic[0]
next_line = schematic[1]

for number in re.finditer(r"\d+", line):

    # Initialize part boolean
    part_number = False

    # Check for adjacent symbols
    if number.start() != 0:
        if line[number.start() - 1] not in no_symbol or \
        set(next_line[number.start() - 1 : number.end()]) - set(no_symbol): 
            part_number = True
    if number.end() != len(line):
        if line[number.end()] not in no_symbol or \
        set(next_line[number.start() : number.end() + 1]) - set(no_symbol): 
                part_number = True

    if part_number:
        part_counter += int(line[number.start() : number.end()])


# Last line
line = schematic[-1]
prev_line = schematic[-2]
for number in re.finditer(r"\d+", line):

    # Initialize part boolean
    part_number = False

    # Check for adjacent symbols
    if number.start() != 0:
        if line[number.start() - 1] not in no_symbol or \
        set(prev_line[number.start() - 1 : number.end()]) - set(no_symbol): 
            part_number = True
    if number.end() != len(line):
        if line[number.end()] not in no_symbol or \
        set(prev_line[number.start() : number.end() + 1]) - set(no_symbol): 
                part_number = True

    # Adjust part_counter
    if part_number:
        part_counter += int(line[number.start() : number.end()])

# Print answer
print(f'The sum of all the part numbers in the engine schematic is {part_counter}.')
```

```python
import regex as re
import numpy as np

# Open input file
with open('input.txt', 'r') as schematic:
    schematic = schematic.read().split("\n")[:-1]

nums = "0123456789"

# Initialize counter
gear_ratio_sum = 0

# All the middle lines
for line_index in range(1, len(schematic) - 1):
    
    # Define current, last and next lines
    line = schematic[line_index]
    prev_line = schematic[line_index - 1]
    next_line = schematic[line_index + 1]

    # Find gears
    for gear in re.finditer(r'\*', line):
        # Find adjacent numbers
        adjacent_nums = []

        for surrounding in [line, prev_line, next_line]:
            for number in re.finditer(r'\d+', surrounding):
                adjacent = False 

                if number.start() != 0: 
                    if gear.start() in range(number.start() - 1, number.end()):
                        adjacent = True
                if number.end() != len(surrounding): 
                    if gear.start() in range(number.start(), number.end() + 1):
                        adjacent = True
        
                if adjacent:
                    adjacent_nums.append(int(surrounding[number.start() : number.end()]))
        
        if len(adjacent_nums) == 2:
            gear_ratio_sum += np.prod(adjacent_nums)

# First and last lines
# First line
line = schematic[0]
next_line = schematic[1]
# Find gears
for gear in re.finditer(r'\*', line):
    # Find adjacent numbers
    adjacent_nums = []

    for surrounding in [line, next_line]:
        for number in re.finditer(r'\d+', surrounding):
            adjacent = False 

            if number.start() != 0: 
                if gear.start() in range(number.start() - 1, number.end()):
                    adjacent = True
            if number.end() != len(surrounding): 
                if gear.start() in range(number.start(), number.end() + 1):
                    adjacent = True
    
            if adjacent:
                adjacent_nums.append(int(surrounding[number.start() : number.end()]))
    
    if len(adjacent_nums) == 2:
        gear_ratio_sum += np.prod(adjacent_nums)

# Last line
line = schematic[-1]
prev_line = schematic[-2]
# Find gears
for gear in re.finditer(r'\*', line):
    # Find adjacent numbers
    adjacent_nums = []

    for surrounding in [line, prev_line]:
        for number in re.finditer(r'\d+', surrounding):
            adjacent = False 

            if number.start() != 0: 
                if gear.start() in range(number.start() - 1, number.end()):
                    adjacent = True
            if number.end() != len(surrounding): 
                if gear.start() in range(number.start(), number.end() + 1):
                    adjacent = True
    
            if adjacent:
                adjacent_nums.append(int(surrounding[number.start() : number.end()]))
    
    if len(adjacent_nums) == 2:
        gear_ratio_sum += np.prod(adjacent_nums)

# Print answer 
print(f"The sum of all the gear ratios in the engine schematic is {gear_ratio_sum}.")
```
